```python
# filter (image shape = 299x299)
img_bgr = cv2.imread(r"Include\prototype_test\28_39.jpeg")
stride = 2
size = 5

def img_filter(img_bgr,size,stride):
    h,w,d = img_bgr.shape
    if img_bgr is not None:
        img_hsv = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2HSV)
        fracture_row = []
        for r in range(h):
            row = []
            for c in range(0,w-size,stride):
                px = img_hsv[r][c:c+size]
                px = np.mean(px, axis=0)
                px = px.tolist()
                row.append(px)
            fracture_row.append(row)

        filtered_img = []
        for r in range(0,h-size,stride):
            filtered_img_row = []
            for c in range(len(fracture_row[0])):
                fracture = []
                for r1 in range(r,r+size):
                    px = fracture_row[r1][c]
                    fracture.append(px)
                fracture = np.mean(fracture, axis=0)
                fracture = fracture.tolist()
                fracture = [round(px) for px in fracture]
                filtered_img_row.append(fracture)
            filtered_img.append(filtered_img_row)

        return filtered_img
                
    elif img_bgr is None:
        print('None')
        return 'None'

    else:
        print('Error')
        return 'Error'
```

```python
# detect and count blobs
# input: BGR image (white background), output: int (blob count)
def blob(bw):
    img = invert(bw)
    #show_img(img)
    b = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    params = cv2.SimpleBlobDetector_Params()
    params.minThreshold = 50
    params.maxThreshold = 200
    params.filterByArea = True
    params.minArea = 30
    params.filterByInertia = True
    params.minInertiaRatio <= 0.5
    params.filterByCircularity = False
    params.filterByConvexity = False

    ver = (cv2.__version__).split('.')
    if int(ver[0]) < 3:
        detector = cv2.SimpleBlobDetector(params)
    else:
        detector = cv2.SimpleBlobDetector_create(params)

    keypoints = detector.detect(b)
    img_with_keypoint = cv2.drawKeypoints(b, keypoints, np.array([]), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    #show_img(img_with_keypoint)

    #print('number = ' + str(len(keypoints)))
    center = []
    for i in keypoints:
        a = []
        x = int(i.pt[0])
        y = int(i.pt[1])
        a.append(y)
        a.append(x)
        center.append(a)
    #print(center)
    #return img_with_keypoint
    return len(keypoints)

img_bgr = cv2.imread(r"Include\prototype_test\28_39.jpeg")
a = mask3(img_bgr)
blob(a)
```

```python
# delete noise and get the blob coordinate
import skimage.io as io
from skimage.measure import regionprops, label
from skimage.color import label2rgb

img_bgr = cv2.imread(r"Include\prototype_test\28_39.jpeg")
bw_gray = bgr2gray(mask3(img_bgr))

# input: gray scale (binary), output: blob coordinate (list of lists)
def blob_extract(bw_gray):
    copy = np.zeros_like(bw_gray)
    labels = label(bw_gray)
    #print(labels[176])
    blob_coords = []
    for r in regionprops(labels):
        if r.area > 20:
            blob_coords.append(r.coords)
            for pixel in r.coords:
                copy[pixel[0], pixel[1]] = 255

    #show_img_gray(copy)
    print('blob number =' + str(len(blob_coords)))
    return blob_coords

blob_extract(bw_gray)
```
