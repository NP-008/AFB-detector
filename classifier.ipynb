```python
import numpy as np
from PIL import Image
import io
import pathlib
import os
import keyboard
import time
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = []
y = []
```

```python
path = 'Include/prototype_test/'
name_lst = os.listdir('Include/prototype_test/')
```

```python
name_lst
```

```python
for img_name in name_lst:
    img = cv2.imread(path + img_name)
    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img2 = Image.fromarray(img1)
    plt.imshow(img2)
    plt.show() 
    key = keyboard.read_key()
    if key == '1':
        print(img_name + ' is negative.')
        x.append(img1)
        y.append('negative')
    elif key == '2':
        print(img_name + ' is positive')
        x.append(img1)
        y.append('positive')
    else:
        print('ERROR')
    img2.close()
    time.sleep(0)
                    
X = np.array(x)
Y = np.array(y)

np.save('X.npy', X)
np.save('Y.npy', Y)
```
